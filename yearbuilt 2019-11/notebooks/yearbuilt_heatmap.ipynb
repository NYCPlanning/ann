```python
import pandas as pd
import csv
import numpy as np
import plotly.graph_objects as go
```

```python
from IPython.core.display import HTML
style = """
<style>
div.output_area {
    overflow-y: scroll;
}
div.output_area img {
    max-width: unset;
}
</style>
"""
HTML(style) 
```

```python
df1 = pd.read_csv("../data/YearBuilt_Heatmap.csv")
df1.head()
```

```python
nta_sorted = df1.sort_values(by=['borocode','ntaname'])
ntaList = nta_sorted['ntaname'].unique().tolist()
ntaList
```

```python
pd.set_option('display.max_rows', None)
decade_sorted = df1.sort_values(by=['decade_built'])
decadeList = decade_sorted['decade_built'].unique().tolist()
decadeList
```

```python
table = pd.pivot_table(df1, values='build_rat', index=['borocode','ntaname'],
                     columns='decade_built', fill_value=0)
table
```

```python
brList = table.values.tolist()
brList
```

```python
fig = go.Figure(data=go.Heatmap(
                   z=brList,
                   x=decadeList,
                   y=ntaList,
                   colorbar=dict(
                    tickfont=dict(size=14, color='#111111')
                   ),
                   colorscale='reds',
                   hoverongaps = False))
fig.update_layout(
    autosize=False,
    width=1500,
    height=2000,
    title="YearBuilt HeatMap by Neighborhood Tabulation Area (NTA)",
    title_x=0.5,
    xaxis_title="Decade",
    yaxis_title="NTA",
    font=dict(
        family="Inconsolata",
        size=7,
        color="#111111"
    ),
    hoverlabel=dict(
        font=dict(family="Inconsolata")
    ),
    titlefont=dict(
        size=21
    ),
    yaxis=dict(autorange="reversed")
)
fig.update_xaxes(title_font=dict(size=14))
fig.update_yaxes(title_font=dict(size=14))

fig.show()
```
